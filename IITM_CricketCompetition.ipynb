```python
import pandas as pd
```

```python
playerListDf = pd.read_csv('playersList.csv')
```

```python
playerListDf
```

```python
playerListAltered = playerListDf.reset_index(level=0)
playerListAltered = playerListAltered.set_index('Players').T.to_dict('list')
playerListAltered['A Nehra'][0]
```

```python
playerListAltered
```

```python
matchDf = pd.read_csv('all_matches.csv',low_memory=False)
```

```python
matchDf.head()
```

```python
matchDfAltered = matchDf.fillna(-1)
```

```python
matchDfAltered['strikerId'] = matchDfAltered['striker'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
matchDfAltered['non_strikerId'] = matchDfAltered['non_striker'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
matchDfAltered['bowlerId'] = matchDfAltered['bowler'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
matchDfAltered['player_dismissedId'] = matchDfAltered['player_dismissed'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
matchDfAltered['other_player_dismissedId'] = matchDfAltered['other_player_dismissed'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
```

```python
matchDfAltered.head()
```

```python
matchDf
```

```python
m = matchDf[(matchDf['ball']<=5.6) & (m['match_id'] == 335982) & (m['innings'] == 1)]
print(m['runs_off_bat'].sum())
print(m['extras'].sum())
print(m['striker'].unique())
print(m['non_striker'].unique())
print(m['bowler'].unique())
```

```python
m2 = matchDf[(matchDf['ball']<=5.6)]

m3 = m2.groupby(['match_id','innings'])['batting_team'].unique().reset_index()
m31 = m2.groupby(['match_id','innings'])['bowling_team'].unique().reset_index()
m4 = m2.groupby(['match_id','innings','striker','bowler']).sum()[['runs_off_bat','extras','wides','noballs','byes','legbyes']].reset_index()
m4.head(10)
```

```python
result = pd.concat([m3,m31, m4], axis=1, join="inner")
result
```

```python
new_df = pd.merge(m3,m31, how='left', left_on=['match_id','innings'], right_on = ['match_id','innings'])
new_df = pd.merge(new_df,m4, how='left', left_on=['match_id','innings'], right_on = ['match_id','innings'])


new_df
```

```python
teamsDict = {}
i = 0
for val in matchDf['batting_team'].unique():
  teamsDict[val] = i
  i = i+1
teamsDict
```

```python
new_df['strikerId'] = new_df['striker'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
new_df['bowlerId'] = new_df['bowler'].apply(lambda x : -1 if x == -1 else playerListAltered[x][0])
new_df['batting_teamId'] = new_df['batting_team'].apply(lambda x : teamsDict[x[0]])
new_df['bowling_teamId'] = new_df['bowling_team'].apply(lambda x : teamsDict[x[0]])
```
